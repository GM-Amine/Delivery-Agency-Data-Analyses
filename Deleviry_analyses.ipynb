```python
df = df.drop(columns=["Station départ","Station déstination"])
```

```python
df = df.drop(columns=["Commercial"])
```

```python
df.head()
```

```python
df["Montant retenu"].unique()
```

```python
df['Montant retenu'] = df['Montant retenu'].replace(['Non configuré', '/', np.nan], 0)
```

```python
df["Montant retenu"].unique()
```

```python
df["Montant retenu"].info()
```

```python
df["Commission sur colis"].unique()
```

```python
df['Commission sur colis'] = df['Commission sur colis'].replace(np.nan, 0)
df["Commission sur colis"].astype(int)
```

```python
df["Commission sur colis"].astype("str")
```

```python
df['Commission sur colis'] = df['Commission sur colis'].astype(str).str.replace('.', '', regex=False).astype(int)
df['Commission sur colis'] = df['Commission sur colis']/10
df['Commission sur colis'] = df['Commission sur colis'].astype(int)
```

```python
df["Commission sur colis"].unique()
```

```python
df["Type de retenu"].astype("category")
```

```python
df["Type de retenu"].unique()
```

```python
df["Surfacturation livreur"].unique()
```

```python
df['Surfacturation livreur'] = (
    df['Surfacturation livreur']
    .astype(str)
    .str.replace('.', '', regex=False)
    .replace('nan', '0')  # This handles string "nan"
    .astype(int)
)
df['Surfacturation livreur'] = df['Surfacturation livreur']/10
df['Surfacturation livreur'] = df['Surfacturation livreur'].astype(int)
```

```python
df["Surfacturation livreur"].unique()
```

```python
df["Extra retenu"].unique()
```

```python
df['Extra retenu'] = (
    df['Extra retenu']
    .astype(str)
    .str.replace('.', '', regex=False)
    .str.replace('/','0', regex=False)
    .replace('nan', '0')  
    .astype(int)
)
df['Extra retenu'] = df['Extra retenu']
df['Extra retenu'] = df['Extra retenu'].astype(int)
```

```python
df["Extra retenu"].unique()
```

```python
df[["Montant retenu","Extra retenu","Commission sur colis"]].info()
```

```python
df[['Montant retenu', 'Extra retenu','Commission sur colis']] = df[['Montant retenu', 'Extra retenu','Commission sur colis']].astype(int)
```

```python
df[['Statut', 'Type de retenu',"Préstation",'Type','Commune',"Wilaya"]] = df[['Statut', 'Type de retenu',"Préstation","Type",'Commune',"Wilaya"]].astype('category')
```

```python
df["Rémunérations livreur"].unique()
```

```python
df['Rémunérations livreur'] =df['Rémunérations livreur'].astype(str).str.replace('Non configuré', '0', regex=False).replace('nan', '0').astype(int)

df['Rémunérations livreur'] = df['Rémunérations livreur'].astype(int)
```

```python
df["Expéditeur"] = df["Expéditeur"].astype("category")
df["Type de retenu livreur"] = df["Type de retenu livreur"].astype("category")
```

```python
df["Station départ"].unique()
```

```python
df["Expédié le"] = pd.to_datetime(df["Expédié le"], format='%d-%m-%Y %H:%M')
```

```python
df["Expédié le"] = pd.to_datetime(df["Expédié le"], dayfirst=True)

# Extract only date (year, month, day)
df["Expédié le"] = df["Expédié le"].dt.date
```

```python
df["Expédié le"] = pd.to_datetime(df["Expédié le"], dayfirst=True)

# Keep only the date part but preserve datetime64 dtype
df["Expédié le"] = df["Expédié le"].dt.floor("D")
```
